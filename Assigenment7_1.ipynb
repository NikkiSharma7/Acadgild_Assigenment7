```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
# 1:

df['FlightNumber']=df['FlightNumber'].fillna(df['FlightNumber'].shift()+10)
df
```

```python
#2:

df[['From','To']]=df['From_To'].str.split('_', expand=True)
df
```

```python
#3
df['From']=df['From'].str.capitalize()
df
```

```python
#4
df.drop('From_To', axis=1, inplace=True)
df
```

```python
#5 
delays = pd.DataFrame()

for i,delaySeries in enumerate(df['RecentDelays'].values):
    for col in range(0,len(delaySeries)):
        delays.at[i,'delay_'+str(col+1)] = delaySeries[col]

```

```python
df = df.drop(columns=['RecentDelays'])

df = pd.concat([df,delays], axis=1)

df
```

```python
s = df.apply(lambda x: pd.Series(x['RecentDelays']),axis=1).stack().reset_index(level=1, drop=True)
s
```
